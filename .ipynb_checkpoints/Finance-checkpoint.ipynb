```python
import pandas as pd

# Cration d'un dataframe de dmo
df = pd.DataFrame({'A': ['X', 'Y', 'Z'], 'B': [1, 2, 3], 'C': [4, 5, 6]})

# Utilisation de la mthode pivot() pour transformer la ligne en colonne
df_pivot = df.pivot(index='A', columns='B', values='C')

print(df_pivot)
```

```python
import pandas as pd

# Cration d'un dataframe de dmo avec des ventes de produits par magasin et par jour
df = pd.DataFrame({'Magasin': ['A','B','A','B','A','B'], 'Jour': ['Lundi','Lundi','Mardi','Mardi','Mercredi','Mercredi'], 'Produit': ['P1','P2','P1','P2','P1','P2'], 'Ventes': [100,200,150,250,50,75]})

# Utilisation de la mthode pivot() pour regrouper les ventes par magasin et par produit 
df_pivot = df.pivot(index='Magasin', columns='Produit', values='Ventes')

print(df_pivot)
```
